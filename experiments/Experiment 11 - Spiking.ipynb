```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn
import pandas
import pytry
import numpy as np
import nengo
```

```python
data = pytry.read('exp11')
data2 = []
for d in data:
    d2 = dict(d)
    d2['rmse'] = d['rmse_in_range'][0]*36/25
    d2['axis'] = 'x'
    data2.append(d2)

    d2 = dict(d)
    d2['rmse'] = d['rmse_in_range'][1]*36/25
    d2['axis'] = 'y'
    data2.append(d2)
    
df = pandas.DataFrame(data2)
```

```python
seaborn.barplot('n_features_1', 'rmse', hue='axis', data=df.query('n_epochs==200'), estimator=np.mean)
plt.xlabel('number of features in first layer')
plt.ylabel('rmse (mm)')
```

```python
rmse_x = np.mean(df.query('n_epochs==200 and axis=="x"')['rmse'])
rmse_y = np.mean(df.query('n_epochs==200 and axis=="y"')['rmse'])

print(rmse_x, rmse_y, np.mean([rmse_x, rmse_y]))


```
